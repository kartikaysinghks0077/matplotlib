```python
import matplotlib.pyplot as plt

# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 30]

# Customize bar properties
bar_width = 0.9
bar_color = 'skyblue'
bar_edgecolor = 'pink'
bar_alpha = 0.8
bar_label = 'Values'

# Create the bar plot
plt.bar(categories, values, width=bar_width, color=bar_color, edgecolor=bar_edgecolor, alpha=bar_alpha, label=bar_label,linewidth=3,align='center',hatch='/')

# Add title and labels
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

```python
'''bar() Function Parameters

x: Positions of the bars (array-like or scalar).

height: Heights of the bars (array-like or scalar).

width: Width of the bars (scalar or array-like, default 0.8).

bottom: Y-coordinates of the bottom of the bars (scalar or array-like, default None).

align: Bar alignment relative to x-coordinates ('center' or 'edge', default 'center').

color: Bar color(s) (str or array-like).

edgecolor: Color of the bar edges (default 'black').

hatch: Pattern for bar hatching (e.g., '/', '\\', '-').

alpha: Transparency level (float, default 1.0).

linewidth: Width of bar edges (float, default None).

label: Label for the bars, used in the legend (str, default None).

tick_label: Custom labels for x-axis ticks (array-like, default None).'''
```

```python
cat = ['A', 'B', 'C', 'D']
val = [10, 20, 15, 30]
bal=[32,43,44,56]
xps=np.arange(len(cat))

plt.figure(figsize=(10,5))
plt.bar(xps+0.2,val,width=0.2,color='g',label='Green vaale ke no.',linewidth=0.9,edgecolor='black',hatch='/')
plt.bar(xps,bal,width=0.2,color='r',label='Red vaale ke no.',linewidth=1.8,edgecolor='blue')

plt.show()
```
